```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

## Point Plots

```python
sns.displot(diamonds['price'])
```

```python
sns.rugplot(diamonds['x'])
```

```python
sns.kdeplot(diamonds['price'])
```

## Distribution Plot

```python
tips = pd.read_csv("187.tips.csv")
```

```python
tips.head()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex',color="#4CB391")
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg',color='r')
```

## Pair Plots

```python
dots = sns.load_dataset('dots')
```

```python
dots.head()
```

```python
sns.pairplot(dots)
```

```python
sns.pairplot(dots,hue='choice')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='day')
```

```python
x = sns.pairplot(tips,hue='day')
x.map_diag(sns.distplot)
x.map_upper(sns.kdeplot)
```

```python
sns.countplot(x='time',data=tips)
```

```python
sns.pointplot(x='total_bill',y='day',data=tips)
```

```python
sns.factorplot(x='smoker',y='tip',data=tips,kind='bar')
```

```python
sns.swarmplot(x='smoker',y='tip',data=tips,hue='sex',palette='magma')
```

```python
sns.stripplot(x='smoker',y='tip',data=tips,hue='sex')
```

```python
sns.barplot(x='day',y='tip',data=tips)
```

```python
sns.boxplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='tip',data=tips,hue='sex')
```
